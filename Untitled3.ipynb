OBS: Para determinar se um sinal é periódico, precisamos verificar se ele se repete em intervalos regulares de tempo.

a) O sinal é periódico.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir o sinal
def x(t):
    return 4 * np.cos(5 * np.pi * t)

# Criar um vetor de tempo
t = np.linspace(0, 6/5, 1000)

# Traçar o sinal
plt.plot(t, x(t))

# Configurar o gráfico
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.title('Sinal x(t) = 4 cos(5π.t)')
plt.grid(True)

# Mostrar o gráfico
plt.show()
```

C)Analisando o sinal, podemos ver que ele é composto por duas partes: uma função degrau e uma função senoidal. A função degrau faz com que o sinal comece em 4 e permaneça constante para t >= 0. A função senoidal adiciona uma oscilação ao sinal, com uma amplitude de 2 e uma frequência de 3 rad/s. O sinal é periódico.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir o sinal
def x(t):
    return 4 * (t >= 0) + 2 * np.sin(3 * t) * (t >= 0)

# Criar um vetor de tempo
t = np.linspace(-3 * np.pi, 6 * np.pi, 1000)

# Traçar o sinal
plt.plot(t, x(t))

# Configurar o gráfico
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.title('Sinal c) x(t) = 4 u(t) + 2 sen(3.t)')
plt.grid(True)

# Mostrar o gráfico
plt.show()
```

h) A equação x[n] = u[n] + p4[n] não é periódica, pois não há um valor de N que satisfaça a equação x[n] = x[n+N] para todos os valores de n. A função u[n] é um degrau unitário, que começa em 0 e aumenta para 1 em n = 0. A função p4[n] é uma função periódica com período N = 4, definida como:
p4[n] = {1, 0, -1, 0} para n = {0, 1, 2, 3} (um período)


```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(-10, 10)  # Cria uma sequência de amostras de -10 a 14
u = np.heaviside(n, 1)  # Função degrau unitário
p4 = np.where(n % 4 == 0, 1, 0)  # Sinal periódico com período 4

x = u + p4  # Soma dos sinais

plt.stem(n, x, label='x[n] = u[n] + p4[n]', linefmt='b-', markerfmt='bo', basefmt=' ', use_line_collection=True)
plt.xlabel('Amostra (n)')
plt.ylabel('Amplitude')
plt.title('Gráfico de x[n] = u[n] + p4[n] para -10 <= n <= 10')
plt.grid(True)
plt.legend()
plt.ylim(-0.2, 2.2)  # Define o intervalo vertical para mostrar a amplitude correta
plt.show()
```

3) Descreva analiticamente os seguintes sinais e trace a derivada de cada um em relação ao tempo

```python
import numpy as np
import matplotlib.pyplot as plt

# Cria um array de tempo de 0 a 6
t = np.linspace(-2, 4, 1000)

# Calcula a derivada da função
dx_dt = -4 * np.pi * np.sin(np.pi * t + np.pi/3)

# Plota o gráfico da derivada
plt.figure(figsize=(8, 4))
plt.plot(t, dx_dt, label='x(t) = −4πsin(πt + π/3)', color='b')
plt.xlabel('Tempo (t)')
plt.ylabel('Amplitude')
plt.title('x(t) = −4πsin(πt + π/3) no Intervalo de -2 a 4')
plt.grid(True)
plt.legend()
plt.ylim(-4*np.pi-1, 4*np.pi+1)  # Define o intervalo vertical para mostrar a amplitude correta
plt.xlim(-2, 4)  # Define o intervalo horizontal de 0 a 6
plt.show()
```
